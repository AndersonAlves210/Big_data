```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo de 2020
df_2020 = pd.read_csv('LeitoOcupacao_2020.csv', sep=';', low_memory=False)

# Remover espaços em branco nas colunas
df_2020.columns = df_2020.columns.str.strip()

# Filtrar os dados para o estado de Pernambuco
df_pe_2020 = df_2020[df_2020['estado'] == 'Pernambuco'].copy()

# Converter as colunas de ocupação para numérico, tratando erros (caso haja valores não numéricos)
ocupacao_colunas = [
    'ocupacaoSuspeitoCli', 'ocupacaoSuspeitoUti', 'ocupacaoConfirmadoCli',
    'ocupacaoConfirmadoUti', 'ocupacaoCovidUti', 'ocupacaoCovidCli',
    'ocupacaoHospitalarUti', 'ocupacaoHospitalarCli'
]

df_pe_2020.loc[:, ocupacao_colunas] = df_pe_2020[ocupacao_colunas].apply(pd.to_numeric, errors='coerce')

# Verificar se a coluna de data está no formato correto. Caso contrário, converter.
df_pe_2020['dataNotificacao'] = pd.to_datetime(df_pe_2020['dataNotificacao'], errors='coerce')

# Criar colunas de mês e ano
df_pe_2020['mes'] = df_pe_2020['dataNotificacao'].dt.month  # Extrair o mês da data
df_pe_2020['ano'] = df_pe_2020['dataNotificacao'].dt.year  # Extrair o ano da data

# Filtrar apenas para o ano de 2020
df_pe_2020 = df_pe_2020[df_pe_2020['ano'] == 2020]

# Criar uma nova coluna que some todas as colunas de ocupação para o total de leitos ocupados
df_pe_2020['total_ocupacao_leitos'] = df_pe_2020[ocupacao_colunas].sum(axis=1)

# Agrupar por mês para calcular a soma da ocupação de leitos
ocupacao_mensal_2020 = df_pe_2020.groupby('mes').agg({
    'total_ocupacao_leitos': 'sum'
}).reset_index()

# Exibir os resultados mensais
print("Ocupação total de leitos por mês em 2020:")
print(ocupacao_mensal_2020)

# Criar gráfico de barras para exibir a ocupação de leitos por mês
plt.figure(figsize=(10, 6))
plt.bar(ocupacao_mensal_2020['mes'], ocupacao_mensal_2020['total_ocupacao_leitos'], color='lightcoral')
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Total de Leitos Ocupados', fontsize=12)
plt.title('Ocupação Total de Leitos por Mês em 2020', fontsize=14)
plt.xticks(ocupacao_mensal_2020['mes'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o arquivo de 2020
df_2020 = pd.read_csv('LeitoOcupacao_2020.csv', sep=';', low_memory=False)

# Remover espaços em branco nas colunas
df_2020.columns = df_2020.columns.str.strip()

# Filtrar os dados para o estado de Pernambuco
df_pe_2020 = df_2020[df_2020['estado'] == 'Pernambuco'].copy()

# Converter as colunas de interesse para numérico, tratando erros
cols_to_convert = [
    'ocupacaoSuspeitoCli', 'ocupacaoSuspeitoUti', 'ocupacaoConfirmadoCli', 'ocupacaoConfirmadoUti',
    'saidaSuspeitaObitos', 'saidaSuspeitaAltas', 'saidaConfirmadaObitos', 'saidaConfirmadaAltas'
]

df_pe_2020[cols_to_convert] = df_pe_2020[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Selecionar apenas as colunas necessárias para a correlação
df_correlation = df_pe_2020[cols_to_convert]

# Calcular a matriz de correlação
correlation_matrix = df_correlation.corr()

# Gerar o gráfico de calor (heatmap) para exibir as correlações
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlação entre Entradas e Saídas de Leitos - 2020')
plt.show()
```

Analisando os pontos mais extremos do gráfico podemos notar que:
- Há uma forte correlação positiva entre as suspeitas de covid em quem estava nos leitos de UTI para a confirmação(0.85), indicando uma forte tendência à infecção se houvesse suspeita da doença.

- Por outro lado há uma baixíssima correlação positiva entre as pessoas que tinham a doença confirmada e recebiam alta com as pessoas que estavam ocupando leitos clínicos e de uti com suspeita da covid(0.05 e 0.19 respectivamente), reforçando o perigo da doença devido sua alta contagiosidade, tendo que parte dos confirmados, ocupar leitos clínicos ou de UTI.

- Outra análise interessante é que a alta correlação positiva entre pessoas que estavam com a confirmação da doença nos leitos clínicos e tiveram que ser realocadas para os leitos de UTI (0.77) o que mostra a rapidez com que a doença evoluia para um caso mais grave.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o arquivo de 2022
df_2022 = pd.read_csv('LeitoOcupacao_2022.csv', delimiter=';', on_bad_lines='skip', low_memory=False)

# Remover espaços em branco nas colunas
df_2022.columns = df_2022.columns.str.strip()

# Filtrar os dados para o estado de Pernambuco
df_pe_2022 = df_2022[df_2022['estado'] == 'Pernambuco'].copy()

# Converter as colunas de interesse para numérico, tratando erros
cols_to_convert = [
    'ocupacaoCovidUti','ocupacaoCovidCli','ocupacaoHospitalarUti','ocupacaoHospitalarCli',
    'saidaSuspeitaObitos', 'saidaSuspeitaAltas', 'saidaConfirmadaObitos', 'saidaConfirmadaAltas'
]

df_pe_2022[cols_to_convert] = df_pe_2022[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Remover linhas onde todas as colunas de interesse são NaN
df_correlation = df_pe_2022[cols_to_convert].dropna(how='all')

# Calcular a matriz de correlação
correlation_matrix = df_correlation.corr()

# Gerar o gráfico de calor (heatmap) para exibir as correlações com uma escala ajustada
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlação entre Entradas e Saídas de Leitos - Pernambuco, 2022')
plt.show()
```

Análise do gráfico:
Ocupação de UTI Covid e Ocupação de Clínica Covid (0,74): Essa forte correlação positiva sugere que o aumento na ocupação de leitos de UTI para Covid está diretamente relacionado ao aumento na ocupação de leitos clínicos para Covid. Isso indica uma pressão simultânea nas duas modalidades de leitos conforme a demanda por atendimento de Covid cresce.

Ocupação de UTI Covid e Ocupação Hospitalar UTI (0,71): Esse valor indica que, conforme a ocupação de leitos de UTI específicos para Covid aumenta, também aumenta a ocupação geral dos leitos de UTI hospitalares. Isso reflete como a pandemia afeta a ocupação total de UTIs, sobrecarregando o sistema como um todo.

Saída Confirmada para Óbitos e Saída Suspeita para Óbitos (0,81): Como mencionado antes, essa forte correlação sugere uma relação entre os óbitos suspeitos e confirmados, possivelmente devido à progressão de pacientes inicialmente suspeitos para casos confirmados de óbito.

Ocupação Hospitalar Clínico e Saída Confirmada para Óbitos (0,03): Essa correlação quase nula indica que a ocupação dos leitos clínicos hospitalares tem pouca ou nenhuma relação com as saídas confirmadas para óbitos.

Ocupação Hospitalar Clínico e Saída Suspeita para Altas (0,05): Esse valor sugere que não há uma correlação forte entre a ocupação dos leitos clínicos hospitalares e as altas suspeitas, indicando que as altas não dependem diretamente do nível de ocupação clínica hospitalar.

Ocupação Hospitalar Clínico e Ocupação de UTI Covid (0,13): A baixa correlação entre essas variáveis sugere que o uso de leitos de UTI específicos para Covid-19 e o uso de leitos hospitalares clínicos não estão fortemente relacionados, o que pode indicar que essas unidades operam de forma relativamente independente no contexto da pandemia.
